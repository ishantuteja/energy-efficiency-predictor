```python
# Separate input features (X) and target variables (y)

X = df.iloc[:, :-2]   # All columns except last 2 (X1–X8)
y = df.iloc[:, -2:]   # Last 2 columns (Y1, Y2)

# Check shapes
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Split data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 80% train, 20% test
    random_state=42     # for reproducibility
)

# Check shapes
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
```

```python
# Select most important features (optional optimization)

important_features = ['X1', 'X2', 'X4', 'X5']

X_train_imp = X_train[important_features]
X_test_imp = X_test[important_features]

print("Reduced feature shape:", X_train_imp.shape)

#Why this is interesting:
#Less noise
#Faster training
#Sometimes better accuracy
```

```python
#  Train Linear Regression model

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

#Model learned: Relationship between features → heating & cooling
```

```python
#  Predict on test data

y_pred = model.predict(X_test)
```

```python
# Evaluate model performance

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# R2 Score
r2 = r2_score(y_test, y_pred)

# MAE
mae = mean_absolute_error(y_test, y_pred)

# RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("R2 Score:", r2)
print("MAE:", mae)
print("RMSE:", rmse)
```

```python
#  Train Random Forest model

from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=100,   # number of trees
    random_state=42
)

rf_model.fit(X_train, y_train)
```

```python
# Predict using Random Forest

y_pred_rf = rf_model.predict(X_test)
```

```python
#  Evaluate Random Forest performance

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# R2 Score
r2_rf = r2_score(y_test, y_pred_rf)

# MAE
mae_rf = mean_absolute_error(y_test, y_pred_rf)

# RMSE
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))

print("Random Forest R2 Score:", r2_rf)
print("Random Forest MAE:", mae_rf)
print("Random Forest RMSE:", rmse_rf)
```

```python
#  Feature importance

import pandas as pd

importance = rf_model.feature_importances_

feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importance
}).sort_values(by='Importance', ascending=False)

print(feature_importance)
```

```python
# Improved Random Forest model (manual tuning)

rf_model_tuned = RandomForestRegressor(
    n_estimators=300,      # more trees → better learning
    max_depth=10,          # prevent overfitting
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42
)

rf_model_tuned.fit(X_train, y_train)
```

```python
#  Predictions with tuned model

y_pred_tuned = rf_model_tuned.predict(X_test)
```

```python
#  Evaluate tuned model

r2_tuned = r2_score(y_test, y_pred_tuned)
mae_tuned = mean_absolute_error(y_test, y_pred_tuned)
rmse_tuned = np.sqrt(mean_squared_error(y_test, y_pred_tuned))

print("Tuned RF R2:", r2_tuned)
print("Tuned RF MAE:", mae_tuned)
print("Tuned RF RMSE:", rmse_tuned)
```

```python
#  Compare models

print("Linear Regression R2:", r2)
print("Random Forest R2:", r2_rf)
print("Tuned Random Forest R2:", r2_tuned)
```

```python
#Feature importance (final model)

importance = rf_model.feature_importances_

feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importance
}).sort_values(by='Importance', ascending=False)

print(feature_importance)
```
